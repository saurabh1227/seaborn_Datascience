```python
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
#create  some data
rng=np.random.RandomState(0)
x=np.linspace(0,10,500)
y=np.cumsum(rng.randn(500,6),0)
```

```python
#plot the data with matplot
plt.plot(x,y)
plt.legend('ABCDEF',ncol=2,loc='upper left')
```

```python
import seaborn as sns
sns.set()
```

```python
plt.plot(x,y)
plt.legend('ABCDEF',ncol=2,loc='upperleft')
```

```python
#seaborn provide high level commands to create a variety of plot types
#useful for statistical data exploration and even some statistical model fitting.
```

```python
data=np.random.multivariate_normal([0,0],[[5,2],[2,2]],size=2000)
data=pd.DataFrame(data,columns=['x','y'])

for col in 'xy':
  plt.hist(data[col],alpha=0.5)
```

```python
for col in 'xy':
  sns.kdeplot(data[col],shade=True)
```

```python
sns.distplot(data['x'])
sns.distplot(data['y'])
```

```python
sns.kdeplot(data['x'])
sns.kdeplot(data['y'])
```

```python
with sns.axes_style('white'):
  sns.jointplot("x","y",data,kind='kde')
```

```python
with sns.axes_style('white'):
  sns.jointplot("x","y",data,kind='hex')
```
